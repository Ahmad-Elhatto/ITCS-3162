# Preprocessing

```python
# Replacing the nulls for all columns with Object datatype
object_columns = df.select_dtypes(include=['object']).columns
df[object_columns] = df[object_columns].fillna("None")
```

```python
df.isnull().sum()
```

```python
# Replacing the nulls for all columns with Object datatype
object_columns = df.select_dtypes(include=['int', 'float']).columns
df[object_columns] = df[object_columns].fillna(0)
```

```python
df.isnull().sum()
```

```python
# Selecting the columns with nominal data
nominal_features = [ 
    'MSZoning', 'Street', 'Alley', 'LotShape', 'LotConfig',  'Neighborhood', 'Condition1', 
    'Condition2','BldgType', 'RoofStyle', 'RoofMatl', 'Exterior1st', 'Exterior2nd', 
    'MasVnrType', 'Foundation', 'Heating', 'GarageType', 'MiscFeature', 'SaleType', 'SaleCondition'
]

# Selecting the columns with ordinal data
ordinal_features = [
    'HouseStyle', 'LandSlope', 'Utilities', 'LandContour', 'ExterQual', 'ExterCond', 'BsmtQual', 
    'BsmtCond', 'BsmtExposure', 'BsmtFinType1', 'BsmtFinType2', 'HeatingQC', 'KitchenQual', 
    'Functional', 'Fence', 'FireplaceQu', 'GarageFinish', 'GarageQual', 'GarageCond', 
    'PavedDrive', 'PoolQC', 'CentralAir', 'Electrical'
]
```

```python
# Encoding ordinal data
from sklearn.preprocessing import LabelEncoder

# Initialize LabelEncoder
label_encoder = LabelEncoder()

encoded_df = df.copy()

# Apply label encoding to selected columns
for column in ordinal_features:
    encoded_df[column] = label_encoder.fit_transform(encoded_df[column])
```

```python
encoded_df.dtypes
```

```python
# Encoding nominal data
encoded_df = pd.get_dummies(encoded_df, columns=nominal_features)
```

```python
encoded_df.head()
```

```python
# Checking the correlation for the features
corr = encoded_df.corr()
plt.figure(figsize=(100,100))
sns.heatmap(corr, annot=False)
plt.title('Correlation Heatmap')
plt.show()
```

# Modeling

```python
X = encoded_df.drop(columns = ['SalePrice'])
y = encoded_df['SalePrice']
```

```python
# Standardizing the data
from sklearn.preprocessing import StandardScaler

X.drop(columns = ['Id'], inplace=True)

X_scaled = StandardScaler().fit_transform(X)
```

```python
X_scaled = pd.DataFrame(X_scaled, columns = X.columns, index = X.index)
X_scaled.head()
```

```python
# Finding the best features
from sklearn.feature_selection import SelectKBest, f_regression

fs = SelectKBest(score_func = f_regression)

best = fs.fit_transform(X_scaled, y)

features = X_scaled.columns

best_features = features[fs.get_support()]

print(best_features)
```

```python
# Splitting the data into training and testing
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(X_scaled[best_features], y, test_size = 0.3)
```

```python
# Linear Modeling
from sklearn import linear_model

lr = linear_model.LinearRegression()

lr_model = lr.fit(X_train, y_train)
```

```python
# Evaluating the Model
print("Intercept", lr_model.intercept_)
print("Slope", lr_model.coef_[0])
print("Coef of determination",lr_model.score(X_test, y_test))
```

```python
# Predictions Using the Linear Regression Model
y_predict = lr_model.predict(X_test)
```

```python
from sklearn import metrics

mean_abs_err = metrics.mean_absolute_error(y_test, y_predict)
mean_sq_err = metrics.mean_squared_error(y_test, y_predict)
root_mse =  metrics.mean_squared_error(y_test, y_predict, squared = False)
```

```python
print("Mean Absolute Error:", mean_abs_err)
print("Mean Squared Error:", mean_sq_err)
print("Root Mean Squared Error:", root_mse)
```

# Experiment 3 

```python
best_k_num = None

best_model = None

best_coef = None

best_root_mse = 1000000

for i in range(10,60,5):
    fs = SelectKBest(score_func = f_regression, k = i)

    best = fs.fit_transform(X_scaled, y)

    features = X_scaled.columns

    best_features = features[fs.get_support()]
    
    X_train,X_test,y_train,y_test = train_test_split(X_scaled[best_features], y, test_size = 0.3)
    lr_model = lr.fit(X_train, y_train)
    
    y_predict = lr_model.predict(X_test)
    
    root_mse =  metrics.mean_squared_error(y_test, y_predict, squared = False)
    
    if best_root_mse > root_mse:
        best_root_mse = root_mse
        best_k_num = i
        best_model = lr_model
        best_coef = lr_model.score(X_test, y_test)
```

```python
print("Best Number of K:", best_k_num)
print("Best Coef of Determination:", best_coef)
print("Best Root Mean Squared Error:", best_root_mse)
```
